```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
# Define the mean and standard deviation of each model (see corresponding notebooks)
means = [49, 49, 49, 48]
stds = [16, 24, 27, 46]

data = [means, stds]

# Create new dataframe
df_err = pd.DataFrame(data, columns=['Base_Model','Linear_Regression', 'Decision_Tree','XGBoost'])
df_err
```

```python
# Visualize the results

figure, ax = plt.subplots(1, 1)

x = np.arange(4)+1

y = df_err.iloc[0,:]
yerr = df_err.iloc[1,:]

plt.errorbar(x, y, yerr=yerr, linestyle= ' ')
ax.plot(1, 49, "or")
ax.plot(2, 49, "or")
ax.plot(3, 48, "or")
ax.plot(4, 49, "or")

plt.xlim([0,5])
plt.ylabel('Delay [min]')
plt.xticks(rotation=45)
ax.set_xticklabels( (' ', 'Base Model','Linear Regression','Decision Tree','XGBoost',' ') );
```
